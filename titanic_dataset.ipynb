```python
def bar_chart(feature):
    survived = tit[tit['Survived']==1][feature].value_counts()
    dead = tit[tit['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('Survived')
```

```python
bar_chart('Parch')
```

```python
bar_chart('Embarked')
```
